```python
#import pandas to create dataframe
import pandas as pd

```

```python
#import all the module needed for the web scraping
import requests
from bs4 import BeautifulSoup
import numpy as np
import re

#all the list are created to append and created a dictionary to display usinh data frame
h1 = []
h2 = []
h3 = []
cano  = []
title_tag = []
title_len = []
#for meta
desc_ = []
desc_len = []
keywords_ = []
robots_ = []
scicon_ = []
p = []
src = []
d = {}                     #dictionar stores urlsand their h1 tags


    
url = "https://www.matlabsolutions.com/"    
page = requests.get(url)
    
soup = BeautifulSoup(page.content, 'html.parser')
    

#title tag
try:
    title = soup.find("title")
    title_tag.append(title.get_text())
    title_len.append(len(title.get_text()))
except:
    title.append("NA")
   
#description 
try:                    #regrex function is used to extract meta tag with property description
    desc = soup.find(attrs={"name": re.compile(r'Description', re.I)}) 
    desc_.append(desc["content"])
    desc_len.append(len(desc["content"]))
except:
    desc_.append("NA")
    
#keywords
try:
    keywoeds = soup.find(attrs={"name": re.compile(r'keywoeds', re.I)})
    keywords_.append(desc["content"])
    
except:
    keywords_.append("NA")
    
#canonical 
try:
    canonical = soup.find(attrs={"rel": re.compile(r'canonical', re.I)})
    cano.append(canonical["href"])
    
except:
    cano.append("NA")  
    
#robots
try:
    robots = soup.find(attrs={"name": re.compile(r'robots', re.I)})
    robots_.append(robots["content"])
    
except:
    robots_.append("NA")
    
#shortcut icon
try:
    sc = soup.find(attrs={"rel": re.compile(r'shortcut icon', re.I)})
    scicon_.append(sc["href"])
    
except:
    scicon_.append("NA")

#hi tg
try:
    h1_tag = soup.findAll("h1")
    h1 = [txt.get_text().replace("\n", "").strip() for txt in h1_tag]     #because get_text() is not avail.for list
except:
    h1.append("NA")
    
    #h2
try:
    h2_tag = soup.findAll("h2")
    h2 = [txt.get_text().replace("\n", "").strip() for txt in h2_tag]     #because get_text() is not avail.for list
except:
    h2.append("NA")
        
     #h3
try:
    h3_tag = soup.findAll("h3")
    h3 = [txt.get_text().replace("\n", "").strip() for txt in h3_tag]    #because get_text() is not avail.for list
except:
    h3.append("NA")
        


       
d["Url"] = url
d["title"] = title_tag
d["title len"] = title_len
d["desc"] = desc_
d["desc len"] = desc_len
d["keywords"] = keywords_ 
d["canonical tag"] = cano 
d["robots"] = robots_
d["short icon"] = scicon_


```
